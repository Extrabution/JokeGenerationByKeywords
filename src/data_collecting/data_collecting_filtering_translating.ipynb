```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def tokenize(data:str):
    inputs = tokenizer(data, return_tensors="pt", truncation=True, padding=True)
    return inputs
```

```python
tokenized_text = []
for i in text:
    tokenized_text.append(tokenize(i))
print(tokenizer.decode(tokenized_text[0].input_ids[0]))
print(tokenizer.convert_ids_to_tokens(tokenized_text[0].input_ids[0]))
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)
```

```python
swear_words_dict = {}
for elem in text:
    if "—" in elem and "\xa0" not in elem and len(elem) < 250:
        data = elem.replace(".", "").replace("1)", "")
        check = data.split("—")
        try:
            word, desc_word = data.split("—")
            desc_words = desc_word.split(",")
            #print(word)
            swear_words_dict[word.strip().lower()] = desc_words
        except Exception as e:
            pass
            #print(f"{e} with data: {check}")

with open("../../data/swear_words_dict.pickle", "wb") as f:
    pickle.dump(swear_words_dict, f)
with open("../../data/swear_words_dict.pickle", "rb") as f:
    swear_words_dict = pickle.load(f)
```

```python
from nltk.tokenize import TweetTokenizer

def nltk_tweet_tokenizer(text: str):
    tk = TweetTokenizer()
    return tk.tokenize(text)
```

```python
import spacy

nlp = spacy.load("ru_core_news_lg")

def lemmatize(text: str):
    token = nlp(text)
    lemmatized_words = []
    for word in token:
        lemmatized_words.append(word.lemma_)
    return lemmatized_words
```

```python
def check_swear(word, i=0):
    if i == 2:
        return word
    neutral_word = word
    if word in swear_words_dict.keys():
        neutral_word = swear_words_dict[word][0]
        neutral_word = check_swear(neutral_word, i)
    return neutral_word

prepared_anecs = []
for anec in anecs:
    prep_anec = anec.split("\n-")
    no_swear_str = ""
    for elem in prep_anec:
        #print(elem)
        words = nltk_tweet_tokenizer(elem)
        for word in words:
            no_swear_str += check_swear(word.lower()) + " "
    prepared_anecs.append(no_swear_str)
prepared_anecs
```

```python
import time
translated_anecs = []
counter = 129 #70
f = open("../../data/translated_anecs.txt", "a", encoding="utf-8")
try:
    for anec in anecs:
        start = time.time()
        #response = requests.post(url=post_url, json={"text":anec, "source":"rus_Cyrl",
        #                                             "target":"eng_Latn" #"token":"686b68039565d64d6c8aaeb171c9af4f6501e3e5"})
        ane = anecs[counter].replace("\n", " ").split(".")
        anes = []
        for i in range(len(ane) - 1):
            ane[i] = ane[i] + "."
        for an in ane:
            a = an.split("?")
            for i in range(len(a)-1):
                a[i] += "?"
            anes += a if len(a)!=0 else an
        anes_s = []
        for an in anes:
            a = an.split("!")
            for i in range(len(a)-1):
                a[i] += "!"
            anes_s += a if len(a) != 0 else an
        translated = ""
        for an in anes_s:
            error_counts = 0
            if an == "":
                continue
            a1_ready, a2_ready = False, False
            while True:
                try:
                    if len(an) > 160:
                        a = an.split(" ")
                        a1 = " ".join(a[:len(a)//2])
                        a2 = " ".join(a[len(a)//2:])
                        if not a1_ready:
                            tr = client.translation(a1, source='rus_Cyrl', target='eng_Latn')
                            translated += tr["translation_text"] + " "
                            a1_ready = True
                        if not a2_ready:
                            tr = client.translation(a2, source='rus_Cyrl', target='eng_Latn')
                            print(f"an: {a1 + a2}, tr: {translated[-len(a1):-1]+tr['translation_text']}")
                            translated += tr["translation_text"] + " "
                            break
                    else:
                        tr = client.translation(an, source='rus_Cyrl', target='eng_Latn')
                        translated += tr["translation_text"] + " "
                        print(f"an: {an}, tr: {tr['translation_text']}")
                        break
                except Exception as e:
                    if error_counts > 15:
                        print("SOME ERROR HAPPENED!!!!, error_counts=16", e)
                        break
                    print("Limit, counter =", counter)
                    time.sleep(10)
                    error_counts += 1
        if error_counts > 15:
            counter += 1
            continue
        #print(a)
        translated_anecs.append(translated)
        f.write(translated + "\n")
        print(time.time()-start, counter)
        counter += 1
except Exception as e:
    print(translated)
    print(e)
finally:
    f.close()
```
