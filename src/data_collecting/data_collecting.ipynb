```python
import requests
from bs4 import BeautifulSoup
response = requests.get("https://www.native-english.ru/jokes")
```

```python
def get_links(html, class_name):
    soup = BeautifulSoup(html, 'html.parser')
    data = soup.find("ul", class_=class_name).find_all("a")
    links = []
    for link in data:
        links.append(link["href"])
    return links
theme_links = get_links(response.text, "list list_big")
theme_links
```

```python
base_url = "https://www.native-english.ru"
anec_links = []
for link in theme_links:
    response = requests.get(base_url+link)
    temp_links = []
    temp_links.append(get_links(response.text, "list list_big"))
    print(temp_links)
    try:
        page_links = get_links(response.text, "pager")
    except:
        continue
    i = 1
    for page in page_links:
        if i == 1:
            i = 10
            continue
        response = requests.get(base_url+page)
        temp_links.append(get_links(response.text, "list list_big"))
    anec_links.append(temp_links)
anec_links
```

```python
anec_links
```

```python
import time
english_anecs = []
start = time.time()
for i in range(len(anec_links)):
    for j in range(len(anec_links[i])):
        for k in range(len(anec_links[i][j])):
            response = requests.get(base_url+anec_links[i][j][k])
            soup = BeautifulSoup(response.text, 'html.parser')
            data = soup.find("div", class_="article article_eng dtext").find("p")
            english_anecs.append(data.text)
    print(i, time.time()-start)
```

```python
import pickle
with open("data/english_anecs_list.pickle", "wb") as f:
    pickle.dump(english_anecs, f)
```
